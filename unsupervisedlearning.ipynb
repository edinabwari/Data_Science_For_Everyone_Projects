```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Fit and transform the Weight column
df['Normalized_Weight'] = scaler.fit_transform(df[['Weight']])

print(df)
```

```python
df=pd.get_dummies(df,columns=['Gender'])
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize the LabelEncoder
LabelEncoder = LabelEncoder()

# Assuming 'Gender' column exists before this step
df['Gender_Female'] = LabelEncoder.fit_transform(df['Gender_Female'])

print(df)

```

```python
from sklearn.preprocessing import LabelEncoder
LabelEncoder = LabelEncoder()
# Assuming 'Gender' column exists before this step
df['Gender_Male'] = LabelEncoder.fit_transform(df['Gender_Male'])

print(df)
```

```python
#saving the preprosed data
df.to_csv('preprocessed_data.csv', index=False)
```
